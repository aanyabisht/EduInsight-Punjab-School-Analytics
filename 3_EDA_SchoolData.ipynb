```python
order = ["I", "II", "III", "IV", "V"]
fig, axs = plt.subplots(1,2, figsize=(11, 4))

sns.barplot(data=df_filtered1, x="Class", y="Kabaddi_Total", order=order , color='g',ax=axs[0])
sns.barplot(data=df_filtered1, x="Class", y="Kabaddi Girls", order=order , color='Purple',ax=axs[0])
axs[0].set_title('No.of Girls kabaddi player')
sns.barplot(data=df_filtered1, x="Class", y="Kabaddi_Total" , color='g',legend=True,ax=axs[1])
sns.barplot(data=df_filtered1, x="Class", y="Kabaddi Boys" , color='blue',legend=True,ax=axs[1])
axs[1].set_title('No.of Boys kabaddi player')
df4 = df[df["Class"].isin(["I", "II", "III","IV","V"])]
kb_counts = df4[["Class", "Kabaddi Boys","Kabaddi Girls"]].set_index('Class')

kb_counts.plot(kind='bar', stacked=True, figsize=(11,3),color={'Kabaddi Boys':'blue','Kabaddi Girls':'yellow'})
sns.despine()
plt.title("Gender Domination in Game")
plt.ylabel("Number of Players")
plt.show()
```
